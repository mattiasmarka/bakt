```python
from matplotlib import pyplot

N = 4
i = 0

fig, subplots = pyplot.subplots(N, 1, figsize=(6.4, 4.8))

def show(title, points):
    global i
    xi = i
    i += 1
    ax = subplots[xi]
    ax.set_title(title)
    ax.get_yaxis().set_visible(False)
    ax.spines['top'].set_visible(False)
    ax.spines['right'].set_visible(False)
    ax.spines['left'].set_visible(False)
    ax.plot(list(points.values()), [[0]] * len(points), "k.")
    for label in points.keys():
        ax.annotate(label, (points[label], 0))

E = -7.882
E_l, E_h = -10, -6

show("Samm 1", {r"$E_l$": E_l, r"$E$": E, r"$E_h$": E_h})

t = 1 / (E_h - E_l)
u = -E * t
u_h = -E_h * t
u_l = -E_l * t
show(r"Samm 2, kus $t=\frac{1}{E_h-E_l}$", {r"$-E_ht$": u_h, r"$-Et$": u, r"-E_lt$": u_l})

Delta = u_l // 1
v_h = u_h - Delta
v_l = u_l - Delta
v = u - Delta
show(r"Samm 3, kus $\Delta=\lfloor-E_lt\rfloor$", {r"$-E_ht-\Delta$": v_h, r"$-Et-\Delta$": v, r"$-E_lt-\Delta$": v_l})

Delta = u_l // 1
w_h = v_h if v_h > 0 else 1 + v_h
w_l = v_l
w = v if v > 0 else 1 + v
show(r"Samm 4, kus $f(x)=x$, kui $x>0$, ja $f(x)=1-x$, kui $x<0$", {r"$f(-E_ht-\Delta$)": w_h, r"$f(-Et-\Delta)$": w, r"$f(-E_lt-\Delta)$": w_l})

fig.tight_layout()
fig.savefig("faasenergiast.jpg")
```
