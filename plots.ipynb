```python
import matplotlib.pyplot as pyplot
import matplotlib

matplotlib.rcParams["figure.dpi"] = 1000

from sim import *

colors = "rgbcmyrgbcmy"

length = 0.74
bound = 1.5

nbits = list(range(1, 9))
ntrots = list(range(1, 4))

energies = []

pyplot.figure()

i = 0
for xntrots in ntrots:
    xenergies = []
    for xnbits in nbits:
        xenergies.append(sim_h2_ground_state(length, xnbits, xntrots, bound, device="CPU", nshots=1024))
    energies.append(xenergies)
    pyplot.plot(nbits, xenergies, "%c--" % colors[i], label="trotterisammude arv = %d" % xntrots)
    i += 1
pyplot.legend()
```
